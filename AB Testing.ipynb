# A/B Тестирование

## Для прохождения игры была выбрана следующая статегия, включающая следующие пункты: 
### Приоритизация задач
Отдавать предпочтение большим релизам для ключевых направлений, которые могут принести максимальный рост выручки:

1. Команда рекомендаций: улучшение персонализированных рекомендаций (повышает конверсию).
2. Команда рекламы: новые рекламные инструменты или улучшение текущих (увеличивает доход от рекламы).
3. Команда поиска: улучшение релевантности поиска (повышает вероятность покупки).
Примечание из условия: большие релизы имеют больший шанс на успех, хотя требуют больше времени.

Старалась ставить задачи средней и низкой сложности для разгрузки команды и быстрого прироста показателей, особенно ближе к концу игры. Это снижает риск упущенного времени.

### Тестирование и запуск
1. Не спешила выпускать эксперименты сразу на всех пользователей. Ждала несколько дней, чтобы убедиться в статистической значимости результатов (снизить вероятность ошибки 1 рода).

2. Минимальный срок тестирования был выбран 3–5 дней, для больших релизов - 7-12 дней.

### Контроль времени
1. Старалась запустить большие релизы для приоритетных команд в первые 60 дней, чтобы  чтобы максимизировать выручку в долгосрочной перспективе.

С такой стратегией на 60 день тестирования картина была следующая:

Большие релизы остались в только у команд Бэка и Дизайна (но видимо бэклог генерируется на рандом и предположение о приоретизации задач было неправильным). В последние 30 дней тестов было принято решение отдать приоритет мелким доработкам: быстрая реализация мелких задач, релиз успешных экспериментов, фокус на финальном приросте.

Результаты были такие: выручка увеличилась примерно на 2%, от исходных значений (по умолчанию 450 000)
